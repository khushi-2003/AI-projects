```python
import cv2
```

```python
path_sun = r"/content/sun.jpg"
```

```python
path_sun
```

```python
img = cv2.imread(path_sun)
img
```

  Computer saves image in RGB format.
  But CV saves image in BGR format

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_rgb
```

```python
path_bw="/content/sun_bw.png"
img_bw = cv2.imread(path_bw)
img_bw
```

```python
img_bw2 = cv2.imread(path_bw,cv2.IMREAD_GRAYSCALE)
img_bw2
```

Visualize using seaborn and matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.imshow(img_bw2)
```

```python
plt.imshow(img_bw2,cmap='gray')
```

```python
sns.heatmap(img_bw2,annot=True,fmt='d')
```
